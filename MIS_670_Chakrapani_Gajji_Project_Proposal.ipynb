# Lyrical Analysis of Taylor Swift’s Discography: Trends, Themes, and Sentiment

**Author:** Chakrapani Gajji

**Date:** November 11, 2025

## Introduction
This project conducts a detailed analysis of Taylor Swift’s discography by combining web scraping with advanced text analytics and sentiment analysis techniques. By systematically collecting data for every song in her catalog—including song titles, album names, album and song links, release year, runtime, streaming play count, and full lyrics—this work blends structured and unstructured data to generate actionable business and artistic insights. The approach includes both basic statistical exploration and deeper textual analysis, focusing on quantifiable patterns and the emotional tone depicted in her lyrics.

## Business and Research Objectives
### Dataset Creation
Build a comprehensive, structured dataset covering song-level and album-level information alongside the full lyrics for each track.

### Quantitative Trends
Analyze the release frequency by year, track song durations, and study play counts to assess career trajectory and listener engagement.

### Text Analytics
Employ techniques such as word clouds, term frequency (TF), inverse document frequency (IDF), and TF-IDF to uncover key words, recurring themes, and distinctive lyrical choices across albums and eras.

### Sentiment Analysis
Assess the emotional valence in lyrics over the course of Taylor Swift’s career, highlighting shifts in sentiment and thematic direction.

## Value Proposition and Stakeholders
This project adds value for a variety of stakeholders, specifically:

- **Record labels and marketers** by identifying effective lyrical content and guiding promotional campaigns;
- **Music producers and songwriters** wishing to learn from historical trends in Taylor Swift’s work;
- **Streaming platforms and digital media analysts** interested in text-driven music categorization and recommendations;
- **Artist management teams** who can use analytical insights for strategic artistic direction;
- **Fans and music critics** looking for a deeper understanding of Taylor Swift’s evolution and songwriting style.

## Expected Benefits
Through the integration of quantitative analysis and text/sentiment analytics, this project will:

- Provide a structured, up-to-date dataset for further musical and textual analysis;
- Illustrate the creative and emotional evolution within Taylor Swift’s extensive catalog;
- Supply summary metrics and visual representations to aid business strategy, artistic development, and academic study;
- Demonstrate the potential of modern web mining and text analytics for innovation in the music industry.

## Data Collection Methodology
All data will be collected using a custom Python web scraper that systematically extracts relevant information from a reputable source, **Spotify**. Data gathering will strictly involve active web scraping, meeting the requirements for originality and practical demonstration of web data extraction.
